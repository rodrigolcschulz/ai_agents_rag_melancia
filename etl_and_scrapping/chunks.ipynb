```python
import os
import glob
import pandas as pd
import matplotlib.pyplot as plt

# Caminho para os arquivos Markdown
caminho_md = r"..\Input\Blog\*.md"

# Listar todos os arquivos .md
arquivos = glob.glob(caminho_md)

print(f"ðŸ”Ž {len(arquivos)} arquivos encontrados.")

# Lista para armazenar dados
dados = []

# Processar cada arquivo
for arquivo in arquivos:
    with open(arquivo, 'r', encoding='utf-8') as f:
        conteudo = f.read()
        palavras = conteudo.split()
        num_palavras = len(palavras)
        dados.append({
            'Arquivo': os.path.basename(arquivo),
            'Palavras': num_palavras
        })

# Criar DataFrame
df = pd.DataFrame(dados)

if not df.empty:
    # EstatÃ­sticas
    estatisticas = df['Palavras'].describe()

    print("\nðŸ“Š EstatÃ­sticas gerais:")
    print(estatisticas)

    # Plotar histograma
    plt.figure(figsize=(10,6))
    plt.hist(df['Palavras'], bins=10, edgecolor='black')
    plt.title('DistribuiÃ§Ã£o do nÃºmero de palavras nos arquivos Markdown')
    plt.xlabel('NÃºmero de Palavras')
    plt.ylabel('FrequÃªncia')
    plt.grid(True)
    plt.show()

    # Exibir a tabela completa
    display(df.sort_values(by="Palavras", ascending=False))
else:
    print("ðŸš« Nenhum arquivo com palavras encontrado ou problema na leitura.")
```
